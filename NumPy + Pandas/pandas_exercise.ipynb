```python
#1. Write a Pandas program to join the two given dataframes along rows and assign all data.
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
        'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
print(pd.concat([student_data1, student_data2]))
```

```python
#2. Write a Pandas program to join the two given dataframes along columns and assign all data.
student_data1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
        'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

student_data2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})
print(pd.concat([student_data1, student_data2], axis=1))
```

```python
#2. Write a Pandas program to calculate all the sighting days of the unidentified flying object (ufo) from current date.
df = pd.read_csv('ufo.csv', low_memory=False)
df.head()
df['Date_time'] = pd.to_datetime(df['Date_time'], errors='coerce')
print(pd.to_datetime('today'))
```

```python
#3. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.
print(pd.to_datetime('today'))
print(df['Date_time'].min())
print((pd.to_datetime('today') - df['Date_time'].min()).days)
```

```python
#4. Write a Pandas program to get all the sighting days of the unidentified flying object (ufo) which are less than or equal to 40 years (365*40 days).
print(df[df['Date_time'] >= pd.to_datetime('today') - pd.DateOffset(days=365*40)])
```

```python
#5. Write a Pandas program to get all the sighting days of the unidentified flying object (ufo) between 1950-10-10 and 1960-10-10.
print(df[(df['Date_time'] >= '1950-10-10 00:00:00') & (df['Date_time'] <= '1960-10-10 00:00:00')])
```

```python
#6. Write a Pandas program to get all the sighting years of the unidentified flying object (ufo) and create the year as column.
df['Year'] = df['Date_time'].dt.year
print(df['Year'])
```

```python
#7. Write a Pandas program to create a plot to present the number of unidentified flying object (UFO) reports per year.
df.Year.value_counts().sort_index().plot(x="Year")
```

```python
#8. Write a Pandas program to extract year, month, day, hour, minute, second and weekday from unidentified flying object (UFO) reporting date.
df['Year'] = df['Date_time'].dt.year
df['Month'] = df['Date_time'].dt.month
df['Day'] = df['Date_time'].dt.day
df['Hour'] = df['Date_time'].dt.hour
df['Minute'] = df['Date_time'].dt.minute
df['Second'] = df['Date_time'].dt.second
df['Weekday'] = df['Date_time'].dt.weekday
print(df[['Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Weekday']])
```

```python
#9. Write a Pandas program to convert given datetime to timestamp.
df['Timestamp'] = df['Date_time'].view('int64')
print(df['Timestamp'])
```

```python
#10. Write a Pandas program to count year-country wise frequency of reporting dates of unidentified flying object(UFO).
result = df.groupby(['Year', 'country']).size()
print(result)
```

# Pandas Plotting

```python
import matplotlib.pyplot as plt
df = pd.read_csv('alphabet_stock_data.csv')
df.head()
```

```python
#1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
plt.figure(figsize=(5,5))
plt.xlabel("Date",fontsize=16, color='black')
plt.ylabel("$ price", fontsize=16, color='black')
df2['Close'].plot(color='green');
plt.show()
```

```python
#2. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-09-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df2 = df.loc[new_df]
plt.figure(figsize=(10,10))
df2.plot(x='Date', y=['Open', 'Close']);
plt.suptitle('Opening/Closing stock prices of Alphabet Inc.,\n 01-04-2020 to 30-09-2020', fontsize=12, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
plt.show()
```

```python
#3. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-4-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
plt.figure(figsize=(6,6))
plt.suptitle('Trading Volume of Alphabet Inc. stock,\n01-04-2020 to 30-04-2020', fontsize=16, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("Trading Volume", fontsize=12, color='black') 
df2['Volume'].plot(kind='bar')
plt.show()
```
